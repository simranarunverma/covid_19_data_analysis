## CONSTRUCTING A DATAFRAME

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
%matplotlib notebook

df = pd.read_csv (r'C:\Users\simra\Downloads\covid_19_india.csv')
print (df)
```

```python
df_latest = df[df['Date']=="25/04/20"]
df_latest.head()
```

## FINDING THE ACTIVE PATIENTS AND TOTAL PATIENTS

```python
df_latest['Total Cases'] = df_latest['Cured'] + df_latest['Deaths'] + df_latest['Confirmed']
df_latest
```

```python
df_latest['Active'] = df_latest['Total Cases'] - df_latest['Cured'] - df_latest['Deaths']
df_latest
```

```python
df_latest.drop(['ConfirmedIndianNational', 'ConfirmedForeignNational'], axis=1)
```

## STATES WITH MAXIMUM AND MINIMUM NUMBER OF ACTIVE CASES

```python
df_latest = df_latest.sort_values(by=['Active'], ascending = False)
plt.figure(figsize=(12,8), dpi=80)
plt.bar(df_latest['State/UnionTerritory'][:5], df_latest['Active'][:5], align='center',color='red')
plt.ylabel('Number of Active Cases', size = 12)
plt.title('States with maximum Active cases', size = 16)
plt.show()
```

```python
df_latest = df_latest.sort_values(by=['Active'], ascending = True)
plt.figure(figsize=(12,8), dpi=80)
plt.bar(df_latest['State/UnionTerritory'][:5], df_latest['Active'][:5], align='center',color='green')
plt.ylabel('Number of Active Cases', size = 12)
plt.title('States with Minimum Active cases', size = 16)
plt.show()
```

## DEATH-TO-CURED RATIO ANALYSIS

```python
df_latest['DTC_RATIO'] = df_latest['Deaths'] / df_latest['Cured'] 
df_latest
```

## PARAMETERS OF DELHI AND MAHARASHTRA

```python
df_maharashtra = df[df['State/UnionTerritory'] == "Maharashtra"]
df_maharashtra.head()
df_maharashtra.reset_index(inplace = True)
df_maha = df_maharashtra.drop(['index', 'Sno', 'Time', 'ConfirmedIndianNational',  'ConfirmedForeignNational'], axis=1)
df_maha
```

```python
df_delhi = df[df['State/UnionTerritory'] == "Delhi"]
df_delhi.head()
df_delhi.reset_index(inplace = True)
df_del = df_delhi.drop(['index', 'Sno', 'Time', 'ConfirmedIndianNational',  'ConfirmedForeignNational'], axis=1)
df_del
```

```python
plt.figure(figsize=(12,5), dpi=80)
plt.plot(df_maha['Date'], df_maha['Deaths'])
plt.plot(df_del['Date'], df_del['Deaths'])
plt.legend(['Maharashtra', 'Delhi'], loc='upper left')
plt.xlabel('Date', size=12)
plt.ylabel('Death Count', size=12)
plt.title('DEATH TOLL COMPARISON', size = 16)
plt.show()
```

```python
plt.figure(figsize=(12,5), dpi=80)
plt.plot(df_maha['Date'], df_maha['Cured'])
plt.plot(df_del['Date'], df_del['Cured'])
plt.legend(['Maharashtra', 'Delhi'], loc='upper left')
plt.xlabel('Date', size=12)
plt.ylabel('Cases Cured', size=12)
plt.title('CURED CASES COMPARISON', size = 16)
plt.show()
```

```python
plt.figure(figsize=(12,5), dpi=80)
plt.plot(df_maha['Date'], df_maha['Confirmed'])
plt.plot(df_del['Date'], df_del['Confirmed'])
plt.legend(['Maharashtra', 'Delhi'], loc='upper left')
plt.xlabel('Date', size=12)
plt.ylabel('Confirmed/Active Cases', size=12)
plt.title('CONFIRMED/ACTIVE CASES COMPARISON', size = 16)
plt.show()
```
